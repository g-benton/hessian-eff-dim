```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
correlations = np.zeros(64) * np.nan
for i in range(1, 65):
    try:
        seed_2 = np.load('../experiments/eigenvalues/hessian_eigs_seed_2/width_'+str(i)+'.npz')['pos_evals']
        seed_1 = np.load('../experiments/eigenvalues/hessian_eigs/width_'+str(i)+'.npz')['pos_evals']
        corr = np.corrcoef(seed_2, seed_1)[0, 1]
        correlations[i] = corr
    except:
        print(i)
```

```python
plt.plot(correlations)
```
