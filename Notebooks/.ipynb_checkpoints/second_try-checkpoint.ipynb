```python
import math
import torch
import hess
import matplotlib.pyplot as plt
from hess.nets import Transformer
```

```python
nx = 500
train_x = torch.linspace(0, 10, nx).unsqueeze(-1)
train_y = torch.sin(train_x * math.pi) + torch.sin(train_x)
```

```python
plt.plot(train_x, train_y)
```

```python
if torch.cuda.is_available():
    torch.cuda.set_device(6)
    train_x, train_y = train_x.cuda(), train_y.cuda()
```

```python
optimus = Transformer(train_x, train_y, n_hidden=2, hidden_size=50,
                     activation=torch.nn.Tanh())
```

```python
optimus = optimus.cuda()
```

```python
optimus.train_net(print_loss=True, lr=0.1,
                 loss_func=torch.nn.MSELoss())
```

```python
out = optimus(train_x)
```
