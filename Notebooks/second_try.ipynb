```python
import math
import torch
import hess
import matplotlib.pyplot as plt
from hess.nets import Transformer
import numpy as np
```

```python
nx = 200
test_x = torch.linspace(0, 10, nx).unsqueeze(-1)
test_y = torch.sin(test_x * 0.5) + torch.sin(test_x) + 0.15*torch.randn_like(test_x)
```

```python
test_x = (test_x - test_x.mean())/test_x.std()
```

```python
splts = [i * nx//5 for i in range(1, 6)]
```

```python
train_x = torch.cat((test_x[:splts[0], :], 
                     test_x[splts[1]:splts[2], :], 
                     test_x[splts[3]:splts[4], :]))
train_y = torch.cat((test_y[:splts[0], :], 
                     test_y[splts[1]:splts[2], :], 
                     test_y[splts[3]:splts[4], :]))
```

```python
plt.plot(test_x.cpu(), test_y.cpu())
plt.scatter(train_x.cpu(), train_y.cpu())
```

```python
use_cuda = torch.cuda.is_available()
if use_cuda:
    torch.cuda.set_device(4)
    train_x, train_y = train_x.cuda(), train_y.cuda()
```

```python
optimus = Transformer(train_x, train_y, n_hidden=5, hidden_size=10,
                     activation=torch.nn.Tanh(), bias=True)
```

```python
if use_cuda:
    optimus = optimus.cuda()
```

```python
optimus.train_net(print_loss=True, lr=0.01, iters=2500,
                 loss_func=torch.nn.MSELoss())
```

```python
out = optimus(train_x)
```

```python
plt.plot(train_x.cpu(), out.detach().cpu())
```

## Let's get this bread

```python
import hess.utils as utils
```

```python
model = optimus.net
par_list = model.parameters()
n_par_w_bias = sum(torch.numel(p) for p in par_list)
```

```python
n_par_w_bias
```

```python
e1 = torch.zeros(n_par_w_bias)
e1[0] = 1
e1 = e1.unsqueeze(0)
```

```python
magic_e1 = utils.unflatten_like(e1, model.parameters())
if use_cuda:
    magic_e1 = [t.cuda() for t in magic_e1]
```

```python
utils.eval_hess_vec_prod(magic_e1, model.parameters(), 
                         net=optimus.net,
                         criterion=torch.nn.MSELoss(), 
                         inputs=train_x, targets=train_y)
```

```python
output = utils.gradtensor_to_tensor(optimus.net, include_bn=True)
```

```python
output.nelement()
```

```python
base_vec = torch.zeros(n_par_w_bias).cuda().unsqueeze(0)
base_vec[0, 0] = 1.

base_vec = utils.unflatten_like(base_vec, model.parameters())
utils.eval_hess_vec_prod(base_vec, model.parameters(),
                        net=optimus.net,
                        criterion=torch.nn.MSELoss(),
                        inputs=train_x, targets=train_y)
output = utils.gradtensor_to_tensor(optimus.net, include_bn=True)

n_par = output.nelement()
```

```python
hessian = torch.zeros(n_par, n_par)
for pp in range(n_par):
    base_vec = torch.zeros(n_par_w_bias).cuda().unsqueeze(0)
    base_vec[0, pp] = 1.
    
    base_vec = utils.unflatten_like(base_vec, model.parameters())
    utils.eval_hess_vec_prod(base_vec, model.parameters(),
                            net=optimus.net,
                            criterion=torch.nn.MSELoss(),
                            inputs=train_x, targets=train_y)
    if pp == 0:
        output = utils.gradtensor_to_tensor(optimus.net, include_bn=True)
        hessian = torch.zeros(output.nelement(), output.nelement())
        hessian[:, pp] = output
    hessian[:, pp] = utils.gradtensor_to_tensor(optimus.net, include_bn=True)
```

```python
e_val, e_vec = np.linalg.eig(hessian)
idx = e_val.argsort()[::-1]   
e_val = torch.FloatTensor(e_val[idx].real)
e_vec = torch.FloatTensor(e_vec[:,idx].real)
```

```python
ne = e_val.div(torch.sum(e_val))
plt.plot(np.cumsum(ne[:10]), marker=".")
```
